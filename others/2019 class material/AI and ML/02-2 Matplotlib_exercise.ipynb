### 1. Áî®ËìùÁ∫øÁªòÂà∂ÂáΩÊï∞ùëì(ùë•) = ùë•^2‚àí2ùë•+3ÁöÑÂõæË°®ÔºåÂπ∂Âú®ÂùêÊ†á(1,2)‰ΩçÁΩÆÂ§ÑÊ†á‰∏ä‰∏Ä‰∏™Á∫¢Ëâ≤ÁöÑÁÇπ


```python
import matplotlib.pyplot as plt
import numpy as np
def f(x):
    result = x * x - 2 * x + 3
    return result
x_list = np.linspace(0,5,num = 100)
# print(x_list)
y = []
for x_item in x_list:
    y.append(f(x_item))
plt.plot(x_list,y,"blue",0,2,"red")
plt.scatter(1,2,c="red")
plt.show()
```

### 2. ‰ΩøÁî®np.linspace()ÂáΩÊï∞ÁîüÊàêt‚àà[0, 2ùúã]„ÄÇÁªôÂÆöùë•=16*sin‚Å°(ùë°)^3Âíåùë¶=13*cos‚Å°(ùë°)‚àí5*cos‚Å°(2ùë°)‚àí2*cos‚Å°(3ùë°)‚àícos‚Å°(4ùë°)ÔºåÁîªÂá∫x-yÂõæË°®ÔºåÂπ∂ÁªôÂõæÊ†áÊ∑ªÂä†‰∏Ä‰∏™È¢òÁõÆÔºöHeart

```python
import matplotlib.pyplot as plt
import numpy as np
import math
def function_x(t):
    result = 16 * (math.sin(t))**3
    return result
def function_y(t):
    result = 13 * math.cos(t) - 5 * math.cos(2*t) - 2 * math.cos(3*t) - math.cos(4*t)
    return result
t_list = np.linspace(0,2*math.pi,num = 1000)
x = []
y = []
for t_item in t_list:
    x.append(function_x(t_item))
    y.append(function_y(t_item))
plt.plot(x,y)
plt.title("X and Y")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
